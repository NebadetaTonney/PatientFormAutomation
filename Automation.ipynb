```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import logging
import time
from selenium.webdriver.support.ui import Select


# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Path to the CSV file and ChromeDriver
csv_file_path = r"C:\Users\tanum\Downloads\edited dataset.csv"
data = pd.read_csv(csv_file_path)
service = Service(executable_path=r'D:\chromedriver-win64\chromedriver.exe')

# Chrome options
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# URLs
login_url = 'https://staging-scweb.arcapps.org/'
form_url = 'https://staging-scweb.arcapps.org/vitals'
facility_url= 'https://staging-scweb.arcapps.org/select-facility'
search_url= 'https://staging-scweb.arcapps.org/client-search'
vital_url= 'https://staging-scweb.arcapps.org/service-points'
vitals_url= 'https://staging-scweb.arcapps.org/vitals'


def login(row):
    driver.get(login_url)
    time.sleep(10)  # Allowing some time for the page to load
    logger.info("Login page loaded.")

    # Log the page source for debugging
    page_source = driver.page_source
    logger.debug(f"Page source: {page_source}")

    # Check for iframes
    iframes = driver.find_elements(By.TAG_NAME, 'iframe')
    logger.info(f"Found {len(iframes)} iframes on the login page.")

    try:
        # Switch to iframe if necessary (replace with the correct iframe id or index)
        if iframes:
            driver.switch_to.frame(iframes[0])
            logger.info("Switched to iframe.")

        # Wait for the username input to be present and enter the username
        username_input = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div/div[1]/input')) 
        )
        username_input.send_keys(str(row['UserName']))
        logger.info("Username entered successfully.")
        
        # Wait for the password input to be present and enter the password
        password_input = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div/div[2]/div[2]/input'))
        )
        password_input.send_keys(str(row['Password']))
        logger.info("Password entered successfully.")
        
        # Wait for the sign-in button to be clickable and click it
        sign_in_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div/div[4]/button'))
        )
        sign_in_button.click()
        logger.info("Sign-in button clicked.")
    except Exception as e:
        logger.error(f"Error during login: {e}")
        logger.debug(driver.page_source)  # Log the page source for debugging
        driver.quit()
        raise

    time.sleep(10)

    def facility():
        driver.get(facility_url)
        time.sleep(10)
        logger.info('Page loaded')

        try:
            dropdown1 = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div[1]/div[1]/select'))
            )
            select1 = Select(dropdown1)
            select1.select_by_visible_text('Lusaka')  # Replace 'Option Text' with the actual option you want to select
            logger.info("Dropdown option selected successfully.")

            # Dropdown 2
            dropdown2 = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div[1]/div[2]/select'))  # Replace with actual ID or locator
            )
            select2 = Select(dropdown2)
            select2.select_by_visible_text('Lusaka')  # Replace 'Option Text 2' with the actual option you want to select

            # Dropdown 3
            dropdown3 = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div[1]/div[3]/div/div[2]/input'))  # Replace with actual ID or locator
            )
            select3 = Select(dropdown3)
            select3.select_by_visible_text('Dr Watson Dental Clinic')  # Replace 'Option Text 3' with the actual option you want to select

            submit_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div/div[3]/form/div[3]/button'))
            )
            submit_button.click()
            logger.info("Form submitted successfully.")

        except Exception as e:
            logger.error(f"Error during form filling: {e}")
            logger.debug("Form error page source saved to form_error_page_source.html")
            driver.quit()
            raise

        time.sleep(10)

        def search(row):
            driver.get(search_url)
            time.sleep(10)
            logger.info('Page loaded')

            try:
                userselect_nrc = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div[2]/div[1]/button[2]')) 
        )
                userselect_nrc.click()
                nrc_input = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div[2]/form/div[1]/div/div/input')) 
        )
                nrc_input.send_keys(str(row['NRC']))

                search_btn= WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div[2]/form/div[2]/button')) 
            )
                search_btn.click()

                attend_patient= WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div[2]/div[2]/div/div[3]/div[2]/div/div/div[2]/div[2]/button[4]')) 
            )
                attend_patient.click()
            
            except Exception as e:
                logger.error(f"Error during login: {e}")
                logger.debug(driver.page_source)  # Log the page source for debugging
                driver.quit()
                raise

        time.sleep(10)


        def vital():
            driver.get(search_url)
            time.sleep(10)
            logger.info('Page loaded')
            try:
                click_vital = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH,'//*[@id="root"]/div[2]/div[2]/div/div/div[2]/div/div/a/div')) 
        )
                click_vital.click()
            except Exception as e:
                logger.error(f"Error during login: {e}")
                logger.debug(driver.page_source)  # Log the page source for debugging
                driver.quit()
                raise
        time.sleep(5)

        def add_vital():
            driver.get(vitals_url)
            time.sleep(10)
            logger.info('Page loaded')
            try:
                add_vital = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.XPATH,'//*[@id="root"]/div[2]/div[2]/div/div[2]/div/div[2]/div[1]/div/div/div/div/div[1]/div/button')) 
        )
                add_vital.click()
            except Exception as e:
                logger.error(f"Error during login: {e}")
                logger.debug(driver.page_source)  # Log the page source for debugging
                driver.quit()
                raise
        time.sleep(5)


        def fill_form(row):
            driver.get(form_url)
            time.sleep(5)  # Allowing some time for the page to load
            logger.info("Form page loaded.")

            try:
                # Wait for the weight input to be present and enter the weight
                weight_input = WebDriverWait(driver, 30).until(
                    EC.presence_of_element_located((By.NAME, 'weight'))
                )
                weight_input.send_keys(str(row['Weight']))
                logger.info("Weight entered successfully.")
                
                # Enter other form fields similarly
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'height'))).send_keys(str(row['Height']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'temperature'))).send_keys(str(row['Temperature']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'systolic'))).send_keys(str(row['Systolic']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'diastolic'))).send_keys(str(row['Diastolic']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'pulseRate'))).send_keys(str(row['PulseRate']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'respiratoryRate'))).send_keys(str(row['RespiratoryRate']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'oxygenSaturation'))).send_keys(str(row['OxygenSaturation']))
                WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.NAME, 'abdominalCircumference'))).send_keys(str(row['AbdominalCircumference']))

                # Wait for the submit button to be clickable and click it
                submit_button = WebDriverWait(driver, 30).until(
                    EC.element_to_be_clickable((By.NAME, 'main_btn'))
                )
                submit_button.click()
                logger.info("Form submitted successfully.")
            except Exception as e:
                logger.error(f"Error during form filling: {e}")
                logger.debug(driver.page_source)  # Log the page source for debugging
                driver.quit()
                raise

        if __name__ == "__main__":
            try:
                for index, row in data.iterrows():
                    logger.info(f"Processing row {index+1}/{len(data)}")
                    login(row)
                    fill_form(row)
            finally:
                driver.quit()





```
